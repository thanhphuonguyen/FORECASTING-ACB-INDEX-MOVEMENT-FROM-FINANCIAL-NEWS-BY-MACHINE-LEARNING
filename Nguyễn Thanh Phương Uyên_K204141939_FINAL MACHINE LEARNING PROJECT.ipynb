	Simple Moving Average (SMA): This technical indicator represents the average closing prices over a specific period. SMA is designed to help traders and investors easily identify the future trends of stocks in the near term. The author uses SMA in combination with other technical tools and indicators to make more accurate investment decisions. Common time periods for this indicator are typically 7 and 25 days for the short term, 50 days for the medium term, and 200 and 300 days for the long term. However, the author prioritizes a 7-day SMA for short-term investment strategies. 

	SMA = 𝐶𝑙𝑜𝑠𝑒 𝑝𝑟𝑖𝑐𝑒1 + 𝐶𝑙𝑜𝑠𝑒 𝑝𝑟𝑖𝑐𝑒2 + ⋯ + 𝐶𝑙𝑜𝑠𝑒 𝑝𝑟𝑖𝑐𝑒𝑛 𝑁𝑢𝑚𝑏𝑒𝑟 𝑜𝑓 𝑡𝑟𝑎𝑑𝑖𝑛𝑔 𝑠𝑒𝑠𝑠𝑖𝑜𝑛 

	Exponential Moving Average (EMA): The EMA is particularly useful for identifying short-term trends and capturing market volatility. It minimizes price noise and allows traders to focus on significant price movements. Due to its faster response to price fluctuations, the EMA is considered more sensitive to short-term market dynamics and can provide timely trading signals.

	EMA𝑡 = 𝐶𝑙𝑜𝑠𝑒 𝑝𝑟𝑖𝑐𝑒𝑡 × 𝐾 + 𝐸𝑀𝐴𝑡−1 × (1 − 𝐾) 

		In which: 
	• Close price t is the close price of the stock at the period of t
	• K is the smoothing factor with the formula where K = 2 1+𝑁 with N as the number of cycles. 
	• EMAt is the EMA value at the period of t. 
	• EMAt-1 is the EMA value at the period of t-1.

	For the variable measuring the trend of price movement of the ACB-Index, the author uses the "close price" index to calculate the return using the following formula. 

	𝑅𝑒𝑡𝑢𝑟𝑛𝑡 = ( 𝐶𝑙𝑜𝑠𝑒 𝑝𝑟𝑖𝑐𝑒𝑡 − 𝐶𝑙𝑜𝑠𝑒 𝑝𝑟𝑖𝑐𝑒𝑡−1) / ( 𝐶𝑙𝑜𝑠𝑒 𝑝𝑟𝑖𝑐𝑒𝑡−1 )

#### 3. MODEL

```python
target = ['Return']
features =  ['Change','7_days_SMA', '14_days_SMA', '7_days_EMA', '14_days_EMA','Positive','Negative']
print('Target: ', target)
print('Features: ', features)
```

##### A. Logistic Regression

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
# ROC-AUC
def _plot_roc_curve(fpr, tpr, thres, auc):
    plt.figure(figsize = (10, 8))
    plt.plot(fpr, tpr, 'b-', color='darkorange', lw=2, linestyle='--', label='ROC curve (area = %0.2f)'%auc)
    plt.plot([0, 1], [0, 1], '--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.title('ROC Curve')
```

```python
X = df[features].values
y = df[target].values
n_state = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1,  random_state = 42)
```

```python
# Build Logistic regression model
LR_classifier = LogisticRegression(multi_class='multinomial',random_state=n_state)

# Input X_train and y_train values ​​into Logistic regression model
LR_classifier.fit(X_train, y_train.ravel())

# Forecast dependent variable value value
y_pred = LR_classifier.predict(X_test)

print('Confusion matrix:')
print(pd.DataFrame(confusion_matrix(y_test,y_pred)),'\n')

print('Classification report:')
print(classification_report(y_test,y_pred))

print('Logistic Regression accuracy: ', round(accuracy_score(y_test, y_pred),4))
```

	Precision is a measure of how accurate the positive predictions are. In this case, the precision for class 0 (label 0) is 0.71, indicating that 71% of the instances predicted as class 0 are actually class 0. The precision for class 1 (label 1) is 0.67, suggesting that 67% of the instances predicted as class 1 are actually class 1

```python
# ROC
y_pred_prob_test = LR_classifier.predict_proba(X_test)[:, 1]
fpr, tpr, thres = roc_curve(y_test, y_pred_prob_test)

roc_auc = auc(fpr, tpr)

_plot_roc_curve(fpr, tpr, thres, roc_auc)
```

	Based on the value of AUC (0.77), the ROC curve of the Logistic Regression model can be considered positive. An AUC value within the range of 0.7 to 0.8 is generally regarded as a relatively good level of classification. The model has the ability to correctly classify cases in both positive and negative classes, although it may not achieve high accuracy or reliability.

##### B. Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import train_test_split

# Split data set into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Build SVM model
SVM_classifier = SVC(kernel = 'linear', random_state = 0)

# Input X_train and y_train values into SVM model
SVM_classifier.fit(X_train, y_train.ravel())

# Predict the value of dependent variable on the training set
y_pred_train = SVM_classifier.predict(X_train)

# Evaluate the results on the training set
print('Confusion matrix on training set:')
print(pd.DataFrame(confusion_matrix(y_train,y_pred_train)),'\n')
print('Classification report on training set:')
print(classification_report(y_train,y_pred_train))
print('SVM accuracy on training set: ', accuracy_score(y_train, y_pred_train))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(SVM_classifier, X_train, y_train)
```

	Based on the confusion matrix and classification report on the training set, the SVM model achieves an accuracy of approximately 0.69 for train. The precision and recall scores for both classes are relatively balanced, indicating that the model performs reasonably well in classifying instances from both the positive and negative classes. However, the overall F1-score is slightly lower, suggesting that the model's ability to balance precision and recall may not be optimal.

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(SVM_classifier, X_test, y_test)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Build SVM model
SVM_classifier = SVC(kernel='linear', random_state=0)

# Give X_train and y_train in SVM
SVM_classifier.fit(X_train, y_train.ravel())

# Predict 
y_pred = SVM_classifier.predict(X_test)

print('Confusion matrix:')
print(pd.DataFrame(confusion_matrix(y_test, y_pred)),'\n')

print('Classification report:')
print(classification_report(y_test, y_pred))

print('SVM accuracy: ', accuracy_score(y_test, y_pred))

# Compute the probabilities of the positive class
svm_probs = SVM_classifier.decision_function(X_test)

# Compute the false positive rate, true positive rate, and thresholds
fpr, tpr, thresholds = roc_curve(y_test, svm_probs)

# Compute the Area Under the ROC Curve (AUC)
auc = auc(fpr, tpr)

# Plot the ROC curve
plt.plot(fpr, tpr, label='SVM (AUC = {:.2f})'.format(auc))

# Set labels and title for the plot
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic - SVM')
plt.legend()

# Show the plot
plt.show()
```

	The AUC (Area Under the Curve) is a performance measure for classification models, and a value of 0.8 indicates that the SVM model has a better classification ability than random guessing. The ratio between the true positive rate (TPR) and the false positive rate (FPR) is quite good.

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Create a dummy tuple
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize model KNN
knn = KNeighborsClassifier(n_neighbors=5)

# Training the model KNN
knn.fit(X_train, y_train)

# Predict the probability of the positive class
y_score = knn.predict_proba(X_test)[:, 1]

# Calculate the ROC curve and the area under the curve (AUC)
fpr, tpr, thresholds = roc_curve(y_test, y_score)
roc_auc = auc(fpr, tpr)

# Draw the ROC curve
plt.plot(fpr, tpr, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic (KNN)')
plt.legend(loc="lower right")
plt.show()
```

	The KNN model with an AUC of 0.88 exhibits strong classification performance on the utilized dataset. This indicates its capability to accurately classify positive and negative instances.
